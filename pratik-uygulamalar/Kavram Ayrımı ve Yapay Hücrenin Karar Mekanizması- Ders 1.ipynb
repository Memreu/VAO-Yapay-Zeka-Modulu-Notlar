Bu proje, Ders 1 notlarÄ±nda iÅŸlediÄŸimiz **Yapay Sinir AÄŸlarÄ± (YSA)** ve **Geri YayÄ±lÄ±m (Backpropagation)** konularÄ±nÄ±n en temel yapÄ± taÅŸÄ± olan tekil bir nÃ¶ronun Python ile canlandÄ±rÄ±lmÄ±ÅŸ halidir.



###  Karar NasÄ±l Verilir? (Neden 0 veya 1?)
HÃ¼cremiz kararlarÄ±nÄ± ders notlarÄ±nda belirtilen matematiksel modele gÃ¶re Ã¼Ã§ aÅŸamayla verir :

1.  **Sinyal Toplama**: Gelen her bir bilgi (girdi), kendi Ã¶nem derecesiyle (aÄŸÄ±rlÄ±k/weight) Ã§arpÄ±lÄ±r ve toplanÄ±r.
2.  **EÅŸik KontrolÃ¼**: Elde edilen toplam "sinyal gÃ¼cÃ¼", aktivasyon fonksiyonundan ($f$) geÃ§irilir. 
3.  **AteÅŸleme (0 vs 1)**: 
    * EÄŸer sinyal gÃ¼cÃ¼ **0'dan bÃ¼yÃ¼kse**: HÃ¼cre uyarÄ±lÄ±r ve sonuÃ§ **1 (AteÅŸlendi)** olur.
    * EÄŸer sinyal gÃ¼cÃ¼ **negatifse**: HÃ¼cre tepki vermez ve sonuÃ§ **0 (Sessiz KaldÄ±)** olur.

###  Ã–ÄŸrenme (Learning) Nedir?
BaÅŸlangÄ±Ã§ta aÄŸÄ±rlÄ±klar rastgele veya sÄ±fÄ±r olduÄŸu iÃ§in hÃ¼cre sÃ¼rekli yanlÄ±ÅŸ kararlar verir. Ancak her hatasÄ±nda ÅŸu sÃ¼reci izler:

* **Hata Hesaplama**: OlmasÄ± gereken gerÃ§ek deÄŸer ile kendi tahmini arasÄ±ndaki farkÄ± bulur.
* **Geri YayÄ±lÄ±m (Backpropagation)**: Bu hatayÄ± geriye doÄŸru ileterek aÄŸÄ±rlÄ±klarÄ±nÄ± ($a_1, a_2, \dots$) gÃ¼nceller.
* **Genelleme**: BirkaÃ§ turun sonunda aÄŸÄ±rlÄ±klar Ã¶yle bir noktaya gelir ki, hÃ¼cre artÄ±k daha Ã¶nce gÃ¶rmediÄŸi durumlarda bile doÄŸru kararÄ± vermeye baÅŸlar.

###  Proje BaÄŸlantÄ±sÄ±
Bu uygulama, 1957 yÄ±lÄ±nda geliÅŸtirilen ilk yapay zeka programlarÄ±ndan biri olan **Perceptron**'u simÃ¼le eder. GÃ¼nÃ¼mÃ¼zÃ¼n devasa GPT modellerinin temelinde yatan o meÅŸhur **"Geri YayÄ±lÄ±m" (1986)** algoritmasÄ±nÄ±n en saf ve anlaÅŸÄ±lÄ±r halidir.

```python
import numpy as np

# --- DERS 1: YAPAY SÄ°NÄ°R HÃœCRESÄ° (PERCEPTRON) SÄ°MÃœLASYONU ---

# 1. ADIM: AKTÄ°VASYON FONKSÄ°YONU
# Notlardaki 'f' fonksiyonu. Gelen sinyal 0'dan bÃ¼yÃ¼kse 1 (ateÅŸle), deÄŸilse 0 (sus)[cite: 173].
def aktivasyon_f(x):
    return 1 if x >= 0 else 0

# 2. ADIM: EÄÄ°TÄ°M VERÄ°LERÄ° (MantÄ±ksal 'VE' KapÄ±sÄ±)
# Ä°ki girdi (x1, x2) ve beklenen doÄŸru sonuÃ§ (y_dogru).
# Sadece iki girdi de 1 ise sonuÃ§ 1 olmalÄ±.
egitim_verisi = [
    (np.array([1, 1]), 1),
    (np.array([1, 0]), 0),
    (np.array([0, 1]), 0),
    (np.array([0, 0]), 0),
]

# 3. ADIM: BAÅLANGIÃ‡ AYARLARI
# AÄŸÄ±rlÄ±klar (a1, a2) ve bias baÅŸlangÄ±Ã§ta tamamen rastgele[cite: 168, 169].
weights = np.array([0.0, 0.0])
bias = 0.0
ogrenme_hizi = 0.1 # Hatadan ne kadar ders Ã§Ä±karacak?

print("--- Yapay HÃ¼cre EÄŸitimi BaÅŸlÄ±yor ---")

# 4. ADIM: Ã–ÄRENME DÃ–NGÃœSÃœ (Geri YayÄ±lÄ±m MantÄ±ÄŸÄ±) [cite: 181, 182, 194]
for epoch in range(20): 
    toplam_hata = 0
    for x, y_dogru in egitim_verisi:
        # Ä°leri Besleme: y = f(a1*x1 + a2*x2 + bias) [cite: 175, 197]
        toplam_sinyal = np.dot(x, weights) + bias
        tahmin = aktivasyon_f(toplam_sinyal)
        
        # Hata Hesaplama: Beklenen vs Tahmin edilen [cite: 201]
        hata = y_dogru - tahmin
        toplam_hata += abs(hata)
        
        # Geri YayÄ±lÄ±m: AÄŸÄ±rlÄ±klarÄ± GÃ¼ncelleme [cite: 205, 206]
        # Hata varsa, aÄŸÄ±rlÄ±klarÄ± hatayÄ± kÃ¼Ã§Ã¼ltecek yÃ¶nde dÃ¼zeltiyoruz.
        weights += ogrenme_hizi * hata * x
        bias += ogrenme_hizi * hata
        
    if toplam_hata == 0:
        print(f"Tur {epoch+1}: Harika! HÃ¼cre kuralÄ± tamamen Ã¶ÄŸrendi.")
        break
    else:
        print(f"Tur {epoch+1}: Hala Ã¶ÄŸreniyor... Toplam hata: {toplam_hata}")

# 5. ADIM: TEST AÅAMASI
print("\n--- EÄŸitim Bitti, Hadi HÃ¼creyi Test Edelim ---")
for x, _ in egitim_verisi:
    sonuc = aktivasyon_f(np.dot(x, weights) + bias)
    print(f"Girdi: {x} -> Karar: {sonuc}")
```

Ã–rneÄŸin bir miktar detaylandÄ±rÄ±lmÄ±ÅŸ hali alttadÄ±r.

```python
import numpy as np

# --- 1. ADIM: KARAR MEKANIZMASI (AKTIVASYON FONKSIYONU) ---
# Sinyal gÃ¼cÃ¼ 0 veya daha bÃ¼yÃ¼kse hÃ¼cre 1 (AteÅŸlendi) deÄŸerini Ã¼retir.
# Sinyal gÃ¼cÃ¼ negatifse hÃ¼cre 0 (Sessiz) deÄŸerini Ã¼retir.
def aktivasyon_f(toplam_sinyal):
    if toplam_sinyal >= 0:
        return 1, "Pozitif Sinyal -> HÃœCRE ATEÅLENDÄ° (1)"
    else:
        return 0, "Negatif Sinyal -> HÃœCRE SESSÄ°Z (0)"

# --- 2. ADIM: EÄITIM VERILERI (VE MANTIK KAPISI) ---
# [Girdi 1, Girdi 2] -> Beklenen DoÄŸru SonuÃ§
egitim_verisi = [
    (np.array([1, 1]), 1),
    (np.array([1, 0]), 0),
    (np.array([0, 1]), 0),
    (np.array([0, 0]), 0),
]

# --- 3. ADIM: BAÅLANGIÃ‡ PARAMETRELERI ---
# AÄŸÄ±rlÄ±klar (weights) baÅŸlangÄ±Ã§ta sÄ±fÄ±rdÄ±r; yani sistem henÃ¼z cahildir.
weights = np.array([0.0, 0.0]) 
bias = 0.0          # Karar eÅŸiÄŸi
ogrenme_hizi = 0.1  # Hatalardan ders Ã§Ä±karma katsayÄ±sÄ±

print("-" * 60)
print("YAPAY HÃœCRE EÄITIM SÃœRECI BAÅLIYOR")
print("-" * 60)

# --- 4. ADIM: EÄITIM DÃ–NGÃœSÃœ (BACKPROPAGATION) ---
# Sistem hata yaptÄ±kÃ§a aÄŸÄ±rlÄ±klarÄ±nÄ± gÃ¼ncelleyerek kuralÄ± Ã¶ÄŸrenmeye Ã§alÄ±ÅŸÄ±r.
for epoch in range(15): 
    hata_yapildi_mi = False
    print(f"\nTUR (EPOCH) {epoch+1}")
    
    for x, hedef in egitim_verisi:
        # Matematiksel iÅŸlem: (x1 * w1) + (x2 * w2) + bias
        sinyal = np.dot(x, weights) + bias
        tahmin, aciklama = aktivasyon_f(sinyal)
        
        # Hata KontrolÃ¼ ve Geri YayÄ±lÄ±m (AÄŸÄ±rlÄ±k GÃ¼ncelleme)
        if tahmin != hedef:
            hata_yapildi_mi = True
            eski_weights = weights.copy()
            
            # FormÃ¼l: Yeni AÄŸÄ±rlÄ±k = Eski AÄŸÄ±rlÄ±k + (Ã–ÄŸrenme HÄ±zÄ± * Hata * Girdi)
            fark = hedef - tahmin
            weights += ogrenme_hizi * fark * x
            bias += ogrenme_hizi * fark
            
            print(f"DURUM: Girdi {x} iÃ§in Hedef {hedef} / Tahmin {tahmin} (HatalÄ±)")
            print(f"BILGI: AÄŸÄ±rlÄ±klar gÃ¼ncellendi: {eski_weights} ===> {weights}")
        else:
            print(f"DURUM: Girdi {x} iÃ§in Tahmin {tahmin} (DoÄŸru) - {aciklama}")

    if not hata_yapildi_mi:
        print("\nSONUÃ‡: TÃ¼m ihtimaller doÄŸru Ã¶ÄŸrenildi. EÄŸitim baÅŸarÄ±lÄ±.")
        break

# --- 5. ADIM: FINAL ANALIZI ---
print("\n" + "-" * 60)
print("EÄITIM SONU ANALIZ RAPORU")
print("-" * 60)
print(f"Ã–ÄŸrenilen Final AÄŸÄ±rlÄ±klarÄ±: {weights}")
print(f"Ã–ÄŸrenilen Final Bias: {bias}")
print("\nAnaliz: HÃ¼cre artÄ±k girdilere bakÄ±p aradaki f(x) kuralÄ±nÄ± Ã§Ã¶zdÃ¼.")
print("Bu sistem artÄ±k dÃ¼ÅŸÃ¼nmÃ¼yor, sadece istatistiksel bir tahmin yapÄ±yor.")
```
